```python
def prepare_data():

    data = pd.read_csv('P11.csv', header=None, names='x')
    data = pd.DataFrame(list(data['x'].str)).T

    x, y = data.shape
    
    x_l = pd.DataFrame(['.'] * y, columns=['x_l'])
    x_r = pd.DataFrame(['.'] * y, columns=['x_r'])

    data = pd.concat([x_l, data, x_r], axis=1)
    data.loc[-1]  = ['.'] * (x + 2)
    data.loc[x] = ['.'] * (x + 2)
    data.index = data.index + 1  # shifting index
    data.sort_index(inplace=True) 
    
    return data    
```

```python
start_x, start_y = (1, 1)
end_x, end_y = (data.shape[0] - 2, data.shape[1] - 2)
```

### part 1

```python
def take_seat(data, x, y):
    
    if data.iloc[x, y] == '.':
        return '.'
    
    else:
    
        seats_around = []

        seats_around.append(data.iloc[x-1, y-1])
        seats_around.append(data.iloc[x-1, y])
        seats_around.append(data.iloc[x-1, y+1])
        seats_around.append(data.iloc[x, y-1])
        seats_around.append(data.iloc[x, y+1])
        seats_around.append(data.iloc[x+1, y-1])
        seats_around.append(data.iloc[x+1, y])
        seats_around.append(data.iloc[x+1, y+1])

        taken_around = sum(map(lambda x : x == '#', seats_around))
        
        #print(data.iloc[x, y], taken_around)
        
        if data.iloc[x, y] == 'L' and taken_around == 0:
            return '#'
        
        if data.iloc[x, y] == 'L' and taken_around > 0:
            return 'L'
        
        if data.iloc[x, y] == '#' and taken_around < 4:
            return '#'
        if data.iloc[x, y] == '#' and taken_around >= 4:
            return 'L'
```

```python
data = prepare_data()
```

```python
CHANGES = 99

while CHANGES>0:
    
#for i in range(2):

    data_new = data.copy()

    changes = 0

    for i in range(start_x, end_x+1):
        for j in range(start_y, end_y+1):
            pos_old = data.iloc[i, j]
            pos_new = take_seat(data, i, j)

            if pos_old != pos_new:
                changes += 1

            data_new.iloc[i, j] = take_seat(data, i, j)

    data = data_new.copy()
    
    CHANGES = changes
    
    print(CHANGES)
```
