```python
def multiplyList(myList) :
     
    # Multiply elements one by one
    result = 1
    for x in myList:
         result = result * x 
    return result
```

```python
objects_counter = []

for scenario in [(1, 1), (1, 3), (1, 5), (1, 7), (2, 1)]:
    
    objects = []
    
    d = scenario[0]
    r = scenario[1]
    x = 0
    y = 0

    while True:        
        x += d
        y += r

        if x >= 323:
            break

        object_ = wide.iloc[x, y]

        if object_ == '#':
            objects.append(object_)

    objects_counter.append(len(objects))

print(objects_counter)
multiplyList(objects_counter)
```
