```python
passport = ''
cust_id = 1

codes = ['hcl', 'hgt', 'pid', 'eyr', 'byr', 'ecl', 'iyr' ]
data_split = pd.DataFrame(columns=[codes])

with open('P4.csv') as f:
    lines=f.readlines()
    for i, line in enumerate(lines):
        
        passport = passport + ' ' + line
        
        if line == '\n':
            
            list_ = passport.replace('\n', '').replace(' ', ',').split(',')
                
            for element in list_:
                for code in codes:
                    if code in element:
                        data_split.loc[cust_id, element[:3]] = element[4:]
                       
            passport = ''
            cust_id += 1

data_split
```

```python
def test_pid(row):
    pid = row['pid'][0]
    if pid.isdecimal() and len(pid) == 9:
        return True

def test_byr(row):
    byr = row['byr'][0]
    if byr.isdecimal():
        if int(byr) >= 1920 and int(byr) <= 2020:
            return True

def test_iyr(row):
    iyr = row['iyr'][0]
    if iyr.isdecimal():
        if int(iyr) >= 2010 and int(iyr) <= 2020:
            return True
        
def test_eyr(row):
    eyr = row['eyr'][0]
    if eyr.isdecimal():
        if int(eyr) >= 2020 and int(eyr) <= 2030:
            return True

def test_hgt(row):
    hgt = row['hgt'][0]
    if hgt[-2:] == 'cm':
        if hgt[:-2].isdecimal():
            if int(hgt[:-2]) >= 150 and int(hgt[:-2]) <= 193:
                return True
    elif hgt[-2:] == 'in':
        if hgt[:-2].isdecimal():
            if int(hgt[:-2]) >= 59 and int(hgt[:-2]) <= 76:
                return True        
    
def test_hcl(row):
    hcl = row['hcl'][0]
    if hcl[0] == '#':
        if len(hcl) == 7:
            if set([x for x in list('#a97842') if x.isalpha()]).issubset(['a', 'b', 'c', 'd', 'e', 'f']):
                return True
        
def test_ecl(row):
    ecl = row['ecl'][0]
    if ecl in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']:
        return True

def validate_pesel2(row):
    
    if row.isnull().sum() == 0:
        
        if test_pid(row):
            if test_byr(row):
                if test_iyr(row):
                    if test_eyr(row):
                        if test_hgt(row):
                            if test_hcl(row):
                                if test_ecl(row):
                                                                
                                    return 1            
            

    return 0
```

```python
data_split['test'] = data_split.apply(lambda row: validate_pesel2(row), axis=1)
data_split
```
